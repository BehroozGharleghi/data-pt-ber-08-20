### Matplotlib Refresher

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame()
df['x'] = np.arange(0,10,1)
df['y'] = df['x']**2
print(df)
```

```python
plt.plot(df['x'], df['y'])
```

```python
df.plot('x', 'y')
```

```python
# use seaborn to plot:
sns.lineplot(df['x'], df['y'])
```

```python
# we have many options for seaborn:
help(sns.lmplot)
```

```python
# seaborn will give the regression line: (we need three arguments)
sns.lmplot('x', 'y', data = df) # writing data = may not be necessary. 

```

```python
# when we dont need th eregression line:
sns.lmplot('x', 'y', data = df, fit_reg=False)
```

### real data

```python
data = pd.read_csv('vehicles.csv')
data.head()
```

```python
# boxplot:
data.boxplot(column = 'Combined MPG') # there are bunch of outliers 
```

```python
print(data['Combined MPG'].mean())
print(data['Combined MPG'].median())
```

```python
# in seaborn:

help(sns.boxplot)
```

```python
sns.boxplot('Combined MPG', data=data)
```

```python
# boxplot of two variables:

sns.boxplot(x='Combined MPG', y='Highway MPG', data=data)
```

```python
# to make it prettier:
```

### violin plot

```python
sns.violinplot('Combined MPG', data=data)
```

```python
# two violin plots:
sns.violinplot(x='Combined MPG', y='Highway MPG', data=data)
```

### Histograms

```python
plt.hist(data['Combined MPG'], bins=50)
```

```python
# histogram with seaborn:
sns.distplot(data['Combined MPG'], bins=100)
```

### Bar plot

```python
# bar plot is for gategorical data:

sns.barplot(x= 'Highway MPG', y= 'Fuel Type', data=data)
```

```python
# we can have different estimators in this diagram:
sns.barplot(x= 'Highway MPG', y= 'Fuel Type', data=data, estimator=np.median) # np.sum, np.mean, etc.
```

```python
#to save the figure:
help(plt.savefig)
# save the figure before you call plt.show(), otherwise it save a blank figure.
```

```python
#to make seaborn graphs prettier we use below:
sns.set_palette()
```

```python
# to save a figure: we use saturation just for the color. we should use .get_figure so that we are able to save the 
# figure
fig = sns.barplot(x= 'Highway MPG', y= 'Fuel Type', data=data, estimator=np.median, saturation =0.9).get_figure()
fig.savefig('example.png')
```
