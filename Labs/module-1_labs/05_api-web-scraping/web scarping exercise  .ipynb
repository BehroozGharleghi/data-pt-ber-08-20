```python
#write a function to get url:

def make_soup(url):
    html = requests.get(url).content
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
# or get the url manually:

url = 'https://en.wikipedia.org/wiki/List_of_police_dog_breeds'
html = requests.get(url).content
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup = make_soup('https://en.wikipedia.org/wiki/List_of_police_dog_breeds')
soup.prettify()
```

```python
# to get the lists from wikipedia:
soup.find_all('li')
```

```python
#use list comprehension to extract the text of each items: with the use of regular expressions : 

[i.text for i in soup.find_all('li') if re.match('[A-Z]', i.text)]
```

```python
# to get just the first 20 words:
[i.text for i in soup.find_all('li') if re.match('[A-Z]', i.text)][0:10]
```

```python
# or use this:
dogs = [i.text for i in soup.find_all('li') if re.match('[A-Z]', i.text)]
dogs[0:10]
```

```python
# we can use more commands in our code such as replace:

dogs = [i.text.replace('[', '').replace(' ', '') for i in soup.find_all('li') if re.match('[A-Z]', i.text)]
dogs[0:10]
```

```python
# we can make a data frame out of our list from wikipedia (dogs)

pd.DataFrame(dogs)
```
