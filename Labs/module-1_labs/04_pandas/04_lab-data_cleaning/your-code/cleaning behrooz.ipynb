```python
# number of missing cells in each column: 
null_cols = df.isnull().sum()
null_cols[null_cols >0]
```

```python
# to check the columns with no missing value:
null_cols[null_cols == 0]
```

```python
# the column with more than 10000 missing values:

null_cols[null_cols > 10000]
```

```python
# count of non missing values:
df.count()
```

```python
# to drop column 
df.drop(['mfrCode','evMotor'], axis =1)
```

```python
# to drop all the columns with more than 10000 missing values as a list:
col_missing_10k = list(null_cols[null_cols > 10000].index) # this mean the name of the columns are index actually.
col_missing_10k
# then we can drop them:
df.drop(col_missing_10k, axis =1)  # axis =1 means look for the names horizontally
```

```python
# to see how many missing values this column has:
sum(df['cylinders'].isnull())
```

```python
# to fill all the NaN values with 0 in cylinder column:
df['cylinders']= df['cylinders'].fillna(0)
```

```python
# to check if the command worked, now it has zero missing values:
sum(df['cylinders'].isnull())
```
