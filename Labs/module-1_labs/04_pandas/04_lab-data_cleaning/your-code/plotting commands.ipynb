```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
x= [1, 2, 3, 4, 5]
y = [1,2,3,4,5]
```

```python
plt.plot(x, y, 'yd')
plt.title("this is a plot")
          
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# ploting multiple series:
z = [5,4,3,2,1]

plt.plot(x, y, color = 'magenta')
plt.plot(x, z, color = 'yellow', linewidth = 5)

#add the legend, legend has a list of arguments
plt.legend(['y', 'z']) # since we plot y and then z, therefore y should ccome first, x is not important.
plt.show()
```

```python
#scatterplot:

a = np.random.random(10)
b = np.random.random(10)

plt.scatter(a, b, color='green')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

### pie chart

```python
animals= [ 'cat', 'dog', 'hamster']
values = [ 20, 30, 50]

plt.pie(values, labels = animals, colors = ['yellow', 'blue', 'red'] )
plt.show()

```

### Bar Chart

```python
plt.bar(animals, height = values)

plt.show()
```

```python
# bar horizontal:

plt.barh(animals, width = values)

plt.show()
```
