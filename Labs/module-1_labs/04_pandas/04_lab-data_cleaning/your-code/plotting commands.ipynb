```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
x= [1, 2, 3, 4, 5]
y = [1,2,3,4,5]
```

```python
plt.plot(x, y, 'yd')
plt.title("this is a plot")
          
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.show()
```

```python
# ploting multiple series:
z = [5,4,3,2,1]

plt.plot(x, y, color = 'magenta')
plt.plot(x, z, color = 'yellow', linewidth = 5)

#add the legend, legend has a list of arguments
plt.legend(['y', 'z']) # since we plot y and then z, therefore y should ccome first, x is not important.
plt.show()
```

```python
#scatterplot:

a = np.random.random(10)
b = np.random.random(10)

plt.scatter(a, b, color='green')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```python
animals= [ 'cat', 'dog', 'hamster']
values = [ 20, 30, 50]

plt.pie(values, labels = animals, colors = ['yellow', 'blue', 'red'] )
plt.show()

```

```python
vehicles = pd.read_csv('vehicles.csv') # we can read it directly 
vehicles.head()
```

```python
# to plot from the vehicles dataframe: in order to plot it nicely we first sort it:
vehicles.sort_values(by = ['City MPG', 'Highway MPG'], inplace = True)
plt.plot(vehicles['City MPG'], vehicles['Highway MPG']) # here x is city and y is highway.
plt.show()


```

```python
# we can directly plot from the file:

vehicles.plot( x='City MPG', y='Highway MPG')
plt.show()
```

```python
# average city mpg by year:


grouped = vehicles[['City MPG', 'Year']]. groupby(['Year']).mean()
print(grouped)
```

```python
## to have two columns with an index we use reset_index()


grouped = vehicles[['City MPG', 'Year']]. groupby(['Year']).mean().reset_index()
print(grouped)
# to plot it:
grouped.plot( x='Year', y='City MPG')
plt.show()
```

```python

grouped.plot(grouped['Year'], grouped[ 'City MPG'])
```

```python
## to print average city mpg and average highway mpg:
grouped2 = vehicles[['City MPG', 'Highway MPG', 'Year']].groupby([ 'Year']).mean().reset_index()
grouped2.plot(x='Year', y=['City MPG', 'Highway MPG'])
plt.show()
```

```python
# lets do bar plot:

grouped2.plot.bar(x= 'Year', y=['City MPG', 'Highway MPG'], stacked = True)
plt.show()
```
